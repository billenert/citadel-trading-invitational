```python
import requests
API_KEY = {'X-API-key': ''}
s = requests.Session()
s.headers.update(API_KEY)
resp = s.get('http://localhost:9999/v1/case')
resp
```

```python
def send_order(sym, limitmarket, side, size):
    resp = s.post("http://localhost:9999/v1/orders", params={"ticker": sym, "type": limitmarket, "action": side, "quantity": size})
    if resp.ok:
        print("SENT:", side, sym, size, "@ ")
        print("Order ID:", resp.json()["order_id"])
    else:
        print("FAIL:", side, sym, size, "@", ":", resp.text)

def get_book(ticker):
    return s.get('http://localhost:9999/v1/securities/book', params={'ticker': ticker}).json()

def best_bid(sym):
    x = get_book(sym)["bids"]
    if x:
        return x[0]
    return None

def best_offer(sym):
    x = get_book(sym)["asks"]
    if x:
        return x[0]
    return None
```

```python
import time
```

```python
def thingy(result, CONSTANT_CORR=0.386, CONSTANT_INT=406, CONSTANT_DIFF=7, CONSTANT_DIFF2=3):
    BB_A = best_bid("APPL")
    BO_A = best_offer("APPL")
    if not BB_A:
        return
    if not BO_A:
        return
    BB_A = BB_A["price"]
    BO_A = BO_A["price"]


    BB_O = best_bid("ORNG")
    BO_O = best_offer("ORNG")
    if not BB_O:
        return
    if not BO_O:
        return
    BB_O = BB_O["price"]
    BO_O = BO_O["price"]

    if CONSTANT_INT + CONSTANT_CORR * BO_A > BB_O + CONSTANT_DIFF:
        send_order("ORNG", limitmarket="MARKET", side="BUY", size=100)
        send_order("APPL", limitmarket="MARKET", side="SELL", size=39)
        result -= 100
        return result
    
    if CONSTANT_INT + CONSTANT_CORR * BB_A < BO_O - CONSTANT_DIFF:
        send_order("ORNG", limitmarket="MARKET", side="SELL", size=100)
        send_order("APPL", limitmarket="MARKET", side="BUY", size=39)
        result += 100
        return result

    if result < 0 and CONSTANT_INT + CONSTANT_CORR * BB_A <= BO_O + CONSTANT_DIFF2:
        send_order("ORNG", limitmarket="MARKET", side="BUY", size=100)
        send_order("APPL", limitmarket="MARKET", side="SELL", size=100 * CONSTANT_CORR)
        result += 100

    if result > 0 and CONSTANT_INT + CONSTANT_CORR * BO_A >= BB_O - CONSTANT_DIFF2:
        send_order("ORNG", limitmarket="MARKET", side="SELL", size=100)
        send_order("APPL", limitmarket="MARKET", side="BUY", size=int(100 * CONSTANT_CORR))
        result -= 100
    return result

```

```python
# 0 to 180 and 180 to 360
hm = 0
while True:
    hm = thingy(result=hm, CONSTANT_CORR=0.378, CONSTANT_INT=404)
    time.sleep(0.1)
```

```python
# 540 to 720
hm = 0
while True:
    hm = thingy(result=hm, CONSTANT_CORR=0.47, CONSTANT_INT=354)
    time.sleep(0.1)
```
